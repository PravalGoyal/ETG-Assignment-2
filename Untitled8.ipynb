# Summer OLympics Analysis

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv('summer_olympics.csv')
dt
```

# Q1 - In how many cities Summer Olympics is held so far?

```python
No_of_city=len(dt['City'].unique())
print("No. of Cities : ", +No_of_city)
```

# Q2 - Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold_medal_list=dt[dt['Medal']=='Gold']
data = []
for sport in dt['Sport'].unique():
    data.append([sport,len(dt[dt['Sport']==sport])])
pd.DataFrame(data,columns = ['Sport','Gold']).sort_values(by='Gold', ascending=False).head().plot(x = 'Sport', y = 'Gold',kind='bar' )
    
```

# Q3- Which sport is having most number of medals so far?

```python
Medals=dt.groupby('Sport').Medal.count()
Medals
Medals.sort_values(ascending=False).head().plot(kind='bar')
```

```python
Medals[::5]
```

#  Q.4-Which player has won most number of medals? (Top 5)

```python
data = []
for athlete in dt['Athlete'].unique():
    data.append([athlete, len(dt[dt['Athlete']==athlete])]) 
```

```python
pd.DataFrame(data,columns = ['Athlete','Award_Won']).sort_values(by='Award_Won', ascending=False).head().plot(x = 'Athlete', y = 'Award_Won', kind = 'bar')
```

# 8. Which sport is having most female Gold Medalists? (Top 5)

```python
dt = pd.read_csv('summer_olympics.csv')
dt = dt[dt['Medal']=='Gold']
dt['Gender'].unique()
```

```python
dt = dt[dt['Gender']=='Women']
data = []
for sport in dt['Sport'].unique():
    data.append([sport,len(dt[dt['Sport']==sport])])
```

```python
pd.DataFrame(data,columns = ['Sport','Women_Players']).sort_values(by='Women_Players', ascending=False).head().plot(x = 'Sport', y = 'Women_Players', kind = 'bar')
```
